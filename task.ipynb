```python
import sqlite3
```

```python
#Connect to database
conn = sqlite3.connect('duomenu_baze.db')
cursor = conn.cursor()

#Drop existing tables
cursor.execute("DROP TABLE IF EXISTS AIRCRAFT")
cursor.execute("DROP TABLE IF EXISTS MODEL")
cursor.execute("DROP TABLE IF EXISTS COMPANIES")
cursor.execute("DROP TABLE IF EXISTS COUNTRY_CODES")

#Create MODEL table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS MODEL (
        MDL_AUTO_KEY INTEGER PRIMARY KEY,
        MODEL_NUMBER TEXT NOT NULL,
        DESCRIPTION TEXT
    )
""")

#Create COMPANIES table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS COMPANIES (
        CMP_AUTO_KEY INTEGER PRIMARY KEY,
        COMPANY_NAME TEXT,
        COC_AUTO_KEY INTEGER
    )
""")

#Create COUNTRY_CODES table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS COUNTRY_CODES (
        COC_AUTO_KEY INTEGER PRIMARY KEY,
        CODE TEXT,
        COUNTRY_NAME TEXT,
        SDF_COC_002 TEXT,
        SDF_COC_003 TEXT,
        FOREIGN KEY (COC_AUTO_KEY) REFERENCES COMPANIES(CMP_AUTO_KEY)
    )
""")

#Create AIRCRAFT table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS AIRCRAFT (
        ACT_AUTO_KEY INTEGER PRIMARY KEY,
        MDL_AUTO_KEY INTEGER,
        CMP_OWNER INTEGER,
        TAIL_NUMBER TEXT,
        FOREIGN KEY (MDL_AUTO_KEY) REFERENCES MODEL(MDL_AUTO_KEY),
        FOREIGN KEY (CMP_OWNER) REFERENCES COMPANIES(CMP_AUTO_KEY)
    )
""")

#Commit changes and close connection
conn.commit()
conn.close()

```

```python
# 2.2. Extracts info about aircrafts (TAIL_NUMBER, MODEL_NUMBER, MODEL_DESCRIPTION, OWNER_COMPANY_NAME, COMPANY_COUNTRY_CODE, COMPANY_COUNTRY_NAME).

conn = sqlite3.connect('duomenu_baze.db')
cursor = conn.cursor()
cursor.execute("""SELECT 
        AIRCRAFT.TAIL_NUMBER,
        MODEL.MODEL_NUMBER,
        MODEL.DESCRIPTION AS MODEL_DESCRIPTION,
        COMPANIES.COMPANY_NAME AS OWNER_COMPANY_NAME,
        COUNTRY_CODES.CODE AS COMPANY_COUNTRY_CODE,
        COUNTRY_CODES.COUNTRY_NAME AS COMPANY_COUNTRY_NAME
    FROM AIRCRAFT
    JOIN MODEL ON AIRCRAFT.MDL_AUTO_KEY = MODEL.MDL_AUTO_KEY
    JOIN COMPANIES ON AIRCRAFT.CMP_OWNER = COMPANIES.CMP_AUTO_KEY
    JOIN COUNTRY_CODES ON COMPANIES.COC_AUTO_KEY = COUNTRY_CODES.COC_AUTO_KEY """)


client_data=cursor.fetchall()
for uzsakymas in client_data:
    print(uzsakymas)
```
